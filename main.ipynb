```python
splitted = df['Date'].str.split("/", expand=True)

df['day'] = splitted[1].astype('int')
df['month'] = splitted[0].astype('int')
df['year'] = splitted[2].astype('int')

df.head()
```

```python
df['is_quarter_end'] = np.where(df['month']%3==0,1,0)
# pd.options.display.max_columns = None
# pd.options.display.max_rows = None
# display(df)
```

```python
data_grouped = df.drop(['Date'], axis=1).groupby('year').mean()

plt.subplots(figsize=(20, 10))

for i, col in enumerate(['Open', 'High', 'Low', 'Close']):
    plt.subplot(2, 2, i+1)
    data_grouped[col].plot.bar()
plt.show()
```

```python
df.drop(['Date'], axis=1).groupby('is_quarter_end').mean()
```

```python
df['open-close'] = df['Open'] - df['Close']
df['high-low'] = df['High'] - df['Low']
df['target'] = np.where(df['Close'].shift(-1) > df['Close'], 1, 0)
```

```python
plt.pie(df['target'].value_counts().values,
        labels=[0, 1], autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(5, 5))

sns.heatmap(df.drop(['Date'], axis=1).corr() > 0.9, annot=True, cbar=False)
plt.show()
```

```python
features = df[['open-close', 'high-low', 'is_quarter_end']]
target = df['target']

scaler = StandardScaler()
features = scaler.fit_transform(features)

X_train, X_valid, Y_train, Y_valid = train_test_split(features, target, test_size=0.1, random_state=2022)
print(X_train.shape, Y_valid.shape)
```

```python
models = [LogisticRegression(),
          SVC(kernel='poly', probability=True),
          XGBClassifier()]

for i in range(3):
    models[i].fit(X_train, Y_train)
    
    print(f"{models[i]}: ")
    print(f"Training Acuuracy: ", roc_auc_score(
        Y_train, models[i].predict_proba(X_train)[:, 1]))
    print(f"Validation Accuracy: ", roc_auc_score(
        Y_valid, models[i].predict_proba(X_valid)[:, 1]))
    print()
```

```python
ConfusionMatrixDisplay.from_estimator(models[0], X_valid, Y_valid)
plt.show()
```
